```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Package Purpose
Allow the someone to create enemies easily that are visually different, by giving them different models and animations. With varing enemmy types such are ranged and melee.

# Analysis Aim
To compare the difference between perfomance of people using the editor and the same people doing in programatically. To see which is the most optimal way of creating an enemy and where focus should be put when making a package.

# Method
- Use a script which can be activated through editor to time the sessions.
- Activated by user.
- User whill switch a bool depending on whether they are using the editor or doing it programaticaly.
- Each person will do it through editor and programatically.


# Data
- **Method**:  How they created the enemys, through editor or programatically.
- **TimeTaken**: How long it took them to from start to finish in seconds.

```python
df = pd.read_csv('ComponentTestResults.csv')
df.columns
```

```python
df.groupby('Method')['TimeTaken'].describe()
```

# Visualisation 

```python
avg_times = df.groupby('Method')['TimeTaken'].mean()
avg_times.plot(kind='bar', color=['skyblue', 'salmon'], figsize=(8, 5))

plt.title('Average Time Taken by Method')
plt.ylabel('Average Time (s)')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df.boxplot(column='TimeTaken', by='Method')
plt.title('Time Taken by Method')
plt.suptitle('')
plt.xlabel('Method')
plt.ylabel('Time Taken (s)')
plt.grid(True)
plt.show()
```

# Data Results
We can see in every case the programatic takes substantially longer time to implement and also from the box plot we can see there is a bigger difference in range of how long it takes individuals to complete creating enemys.

# Conclusion
We can see from the results that players  find it much easier using the editor and do so much quicker, possibly due to the visual aspect of it, its a lot easier to foget an assingment when programming if there isn't an empty box. we can also see from this that the range of differences in time is much greater programtically. I believe this to be factored also due to varying skill levels, as most people would be able to complete this little test easily through the editor but programmatically has its own set of additional technical hurdles.
